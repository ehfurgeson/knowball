Could tweets give a signal to player performance? This is an experiment to see if basketball players might perform differently based on their recent tweets, to quantify tweets they will be embedded via BERT or BERTweet, then these embeddings will be utilized in several models to determine if any tweets or patterns of tweeting could be predictive or player performance.

```python
import csv
import activity

def read_csv(file_path) -> list[str]: 
    """Read Twitter handles from a CSV file."""
    twitter_handles = []
    with open(file_path, mode='r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header
        for row in reader:
            twitter_handles.append(row[2])  # Twitter handles are in the third column
    return twitter_handles

handles = read_csv("x.csv")
results = activity.analyze_twitter_users(handles)
```
